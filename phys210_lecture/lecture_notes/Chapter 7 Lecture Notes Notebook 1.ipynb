We will solve integrals by approximation: <br><br>
$$\int_a^bf(x)dx\approx\sum_{i=0}^{n-1}c_if(x_i)\text{, where }
\begin{array}{}
x_i\equiv\text{nodal abscissas}^* \\
c_i\equiv\text{weights}
\end{array}$$ <br>
<sup>* Abscissa: the x-coordinate of a point in a two-dimensional coordinate system. </sup> <br>
via two categories of approximation:
1. Closed Methods: $a$ and $b$ are abscissas.
    * Newton-Cotes methods are closed-method approximations. <br><br>
2. Open Methods: $a$ and $b$ are <i>not</i> abscissas.
    * Gaussian Quadrature is an open-method approximation (sometimes exact!)

<u>Newton-Cotes Methods</u>
* Assumes that an integral can be approximated as a sum of the areas of simple shapes, e.g., rectangles, triangles, trapezoids, etc. <br><br>
* Consider $n$ equally spaced points $x_i$ determined via <br><br>
$$\begin{array}{}
x_i=a+ih, &
i=0,1,...,n-1
\end{array}$$
where $h=\dfrac{b-a}{n-1}$. <br><br>
So $x_0=a$ and $x_{n-1}=b$. <br><br>
* Let $N$ be the number of panels defined by the width between two points: <br><br>
$$N=n-1$$

##### <u>Rectangle Rule:</u>
* For the case of $N=1$,
$$\displaystyle{\int_{x_i}^{x_{i+1}}}f(x)dx\approx hf(x_i)$$ <br>
    - Defining $h$ by the left abscissa, $f(x_i)$, is a <i>left-hand rectangle rule</i> for approximation. <br><br>
    - Defining $h$ by the right abscissa, $f(x_{i+1})$, is a <i>right-hand rectangle rule</i> for approximation. <br><br>
* <i>Composite rectangle rule</i>: <br><br>
    $$\begin{array}{lll}
    \displaystyle{\int_a^b}f(x)dx & = & \displaystyle{\sum_{i=0}^{n-2}\int_{x_i}^{x_{i+1}}}f(x)dx \\
    \; & \approx & \displaystyle{hf(x_0)+hf(x_1)+hf(x_2)+\ldots+hf(x_{n-2})}
    \end{array}$$ <br>
* Error Analysis: Taylor expand $f(x)$ around $x_i$ <br><br>
$$f(x)=f(x_i)+(x-x_i)f'(x_i)+\cdots$$ <br>
stopping at first order, <br><br>
$$f(x)=f(x_i)+(x-x_i)f'(\xi_i)$$ <br>
where $\xi_i$ is a point between $x_i$ and $x$. <br><br>
    - Integrate $f(x)$ by definition of the composite rectangle rule: <br><br>
    $$\int_{x_i}^{x_{i+1}}f(x)dx=\int_{x_i}^{x_{i+1}}{f(x_i)+(x-x_i)f'(\xi_i)}dx$$ <br>
Let $u=\dfrac{x-x_i}{h}$:
$$\begin{array}{}
x=x_i & \Longrightarrow & u=0 & \; &\frac{du}{dx}=\frac{1}{h} \\
x=x_{i+1} & \Longrightarrow & u=1 & \; & dx=hdu
\end{array}$$ <br>
So,
$$\begin{array}{}
\displaystyle{\int_{x_i}^{x_{i+1}}}{f(x_i)+(x-x_i)f'(\xi_i)}dx & = & h\displaystyle{\int_0^1}{f(x_i)+huf'(\xi_i)}du \\
\; & = & hf(x_i)+\dfrac{1}{2}h^2f'(\xi_i)
\end{array}$$ <br>
The absolute error in $\displaystyle{\int_{x_i}^{x_{i+1}}}f(x)dx\approx hf(x_i)$ is then: <br><br>
$$\varepsilon_i=\dfrac{1}{2}h^2f'(\xi_i)$$ <br>
So the total error $\varepsilon$ is
$$\begin{array}{}
\varepsilon & = & \displaystyle{\sum_{i=0}^{n-2}}\varepsilon_i \\
\; & = & \dfrac{1}{2}h^2\displaystyle{\sum_{i=0}^{n-2}}f'(\xi_i)
\end{array}$$ <br>
According to Mean Value Theorem, $\exists \; \xi$ s.t. $f'(\xi)=\frac{\underset{i=0}{\overset{n-2}{\sum}}f'(\xi_i)}{n-1}$. <br><br>
So introducing a $\frac{n-1}{n-1}$ coeffficient to $\varepsilon$: <br><br>
$$\begin{array}{}
\dfrac{n-1}{2}h^2\dfrac{\sum_{i=0}^{n-2}f'(\xi_i)}{n-1} & = & \dfrac{n-1}{2}h^2f'(\xi)
\end{array}$$ <br>
Recall that $h=\dfrac{b-a}{n-1}$: the <i>absolute error</i> in the <i>composite rectangle forumula</i> is <br><br>
$$\varepsilon=\frac{b-a}{2}hf'(\xi)$$ <br>
<sup>note: $\varepsilon$ is not a constant multiple of $h$ since $f'(\xi)$ is dependent on $h$; if this is assumed, then $\varepsilon$ is actually $\mathcal{O}(h)$.</sup>

##### <u>Midpoint Rule:</u>
* For the case of $N=1$ of a rectangle rule: <br><br>
    - $h$ is defined as the same, i.e., the width given by the distance between $x_i$ and $x_{i+1}$. <br><br>
    - If the height $f(x)$ is defined as neither the left or right of the panel but instead a midpoint between the two, then <br><br>
    $$\int_{x_i}^{x_{i+1}}f(x)dx\approx hf\left(x_i+\frac{h}{2}\right)$$ <br>
    Let $\Delta i\equiv x_i+\frac{h}{2}$: expanding $f(x)$ about $\Delta i$ gives us <br><br>
    $$f(x)\approx f(\Delta i)+(x-\Delta i)f'(\Delta i)+\frac{1}{2}(x-\Delta i)^2f''(\xi_i)$$ <br>
    Let $u=\dfrac{x-\Delta i}{\tfrac{h}{2}}$: the absolute error in $\displaystyle{\int_{x_i}^{x_{i+1}}}f(x)dx\approx hf(x_i)$ is then <sup>*</sup> <br><br>
    $$\varepsilon_i=\frac{h^3}{24}f''(\xi_i)$$
<sup>* : work shown in Homework 9 file</sup> <br><br>
So the total error $\varepsilon$ is
$$\varepsilon=\frac{h^3}{24}\sum_{i=0}^{n-2}f''(\xi_i)$$ <br>
And by the same method done in rectangle rule, $\varepsilon$ is finally <br><br>
$$\varepsilon=\frac{b-a}{24}h^2f''(\xi)$$

### Midpoint vs. Lefthand Approximation Rules Comparison for evaluating the integral of $f(x)=\dfrac{1}{\sqrt{1+x^2}}$ on the interval $[0,1]$.

```python
import numpy as np

f = lambda x: 1./np.sqrt(x**2+1)

def params(xs):
    a = xs[0]
    n = xs.shape[0]
    b = xs[n-1]
    h = (b-a)/(n-1)
    return a, b, n, h

def lefthand(f, xs):
    a, b, n, h = params(xs)
    n = xs.shape[0]
    lh = h*f(xs[:-1]).sum() #the [:-1] slicing is important
    return lh

def midpoint(f, xs):
    a, b, n, h = params(xs)
    mp = h*f(xs[:-1]+h/2).sum()
    return mp

anal_int = np.log(1+np.sqrt(2))

ns = [11, 51, 101, 1001, 10001]
for n in ns:
    xs = np.linspace(0.,1.,n)
    mp = midpoint(f, xs)
    lh = lefthand(f, xs)
    print(f"n={n:5d}; mp={mp:7.5f}; lh={lh:7.5f}; analytic = {anal_int:7.5f}")
```

<u>Integration from Interpolation:</u>
* Approximates via linear terms insteads of constants for integratable functions. <br><br>
* One panel with 2 abscissas determines the line from $f(x_i)$ to $f(x_{i+1})$. <br><br>
* We will use Lagrange interpolation: <br><br>
    - Given $q$ data points as the set of $\left(x_{i+j},f(x_{i+j})\right)$ for $j=0,1,\ldots,q-1$, the elementary interval for approximation is <br><br>
    $$\int_{x_i}^{x_{i+q-1}}f(x)dx$$ <br>
    - Approximate the elementary interval using <br><br>
    $$p(x)=\sum_{j=0}^{q-1}f(x_{i+j})L_{i+j}(x)$$ <br>
    - Recall Lagrange Polynomials: since we go from $x_i$ to $x_{i+q-1}$, <br><br>
    $$\begin{array}{}
    L_k(x)=\dfrac{\prod_{j=0,j\neq k}^{n-1}(x-x_j)}{\prod_{j=0,j\neq k}^{n-1}(x_k-x_j)} &
    \Longrightarrow &
    L_{i+j}(x)=\dfrac{\prod_{k=0,k\neq j}^{q-1}(x-x_{i+k})}{\prod_{k=0,k\neq j}^{q-1}(x_{i+j}-x_{i+k})} &
    \text{for } j=0,1,\ldots,q-1
    \end{array}$$ <br>
    - The elementary interval given $p(x)$ is then <br><br>
    $$\begin{array}{}
    \displaystyle{\int_{x_i}^{x_{i+q-1}}}f(x)dx & \approx & \displaystyle{\int_{x_i}^{x_{i+q-1}}}p(x)dx & = &
    \displaystyle{\sum_{j=0}^{q-1}f(x_{i+j})\int_{x_i}^{x_{i+q-1}}L_{i+j}(x)dx} \\
    \; & \; & \; & = &\displaystyle{\sum_{j=0}^{q-1}f(x_{i+j})w_{i+j}}
    \end{array}$$ <br>
    where $w_{i+j}\equiv\displaystyle{\int_{x_i}^{x_{i+q-1}}}L_{i+j}(x)dx$ <sup>*</sup> <br><br>
    <sup>* $L_{i+j}(x)$ does not depend of $f(x)$</sup> <br><br>
    - The weights $w_{i+j}$ can be found for a given $q$ <u>once</u>.

##### <u>Trapezoid Rule:</u>
* Let $q=2$ for the integration interpolation: <br><br>
$$\begin{array}{ll}
L_i(x)=\dfrac{x-x_{i+1}}{\underbrace{x_i-x_{i+1}}_{-h}} & L_{i+1}(x)=\dfrac{x-x_i}{\underbrace{x_{i+1}-x_i}_{h}} \\
j=0 & j=1 \\
k=1 & k=0
\end{array}$$ <br>
* The weights of the interpolating polynomials are: <br><br>
$$\begin{array}{}
w_i & = & \displaystyle{\int_{x_i}^{x_{i+1}}}L_i(x)dx & = & \displaystyle{-\frac{1}{h}\int_{x_i}^{x_{i+1}}}(x-x_{i+1})dx \\
\; & \; & \; & = & -\dfrac{1}{h}\left[\dfrac{1}{2}x^2-x_{i+1}\cdot x\right]_{x_i}^{x_i+h \;^*} \\
\; & \; & \; & = & -\dfrac{1}{h}\left[\dfrac{1}{2}(x_i+h)^2-\dfrac{1}{2}x_i^2-x_{i+1}h\right] \\
\; & \; & \; & = & -\dfrac{1}{h}\left[x_ih+\dfrac{1}{2}h^2-x_{i+1}h\right] \\
\; & \; & \; & = & -\dfrac{1}{h}\left[h(-h)+\dfrac{1}{2}h^2\right] \\
\; & \; & \; & = & \dfrac{h}{2}
\end{array}$$ <br>
$$\begin{array}{}
w_{i+1} & = & \displaystyle{\int_{x_i}^{x_{i+1}}}L_{i+1}(x)dx & = & \displaystyle{\frac{1}{h}\int_{x_i}^{x_{i+1}}}(x-x_i)dx \\
\; & \; & \; & = & \dfrac{1}{h}\left[\dfrac{1}{2}x^2-x_i\cdot x\right]_{x_i}^{x_i+h} \\
\; & \; & \; & = & \dfrac{1}{h}\left[\dfrac{1}{2}(x_i+h)^2-\dfrac{1}{2}x_i^2-x_ih\right] \\
\; & \; & \; & = & \dfrac{1}{h}\left[x_ih+\dfrac{1}{2}h^2-x_ih\right] \\
\; & \; & \; & = & \dfrac{1}{h}\left[\dfrac{1}{2}h^2\right] \\
\; & \; & \; & = & \dfrac{h}{2}
\end{array}$$ <br>
<sup>* : $x_{i+1}=x_i+h$</sup>
* The composite Trapezoid rule is then: <br><br>
$$\begin{array}{}
\displaystyle{\int_{x_i}^{x_{i+1}}}f(x)dx & \approx & \displaystyle{\sum_{j=0}^{q-1}}w_{i+j}f(x_{i+j}) \; = \; \dfrac{h}{2}\big[f(x_i)+f(x_{i+1})\big] \\
\displaystyle{\int_a^b}f(x)dx & = & \displaystyle{\sum_{i=0}^{n-2}\int_{x_i}^{x_{i+1}}}f(x)dx \\
\; & \approx & \dfrac{h}{2}\big[f(x_0)+f(x_1)+f(x_1)+f(x_2)+f(x_2)+f(x_3)+\ldots+f(x_{n-2})+f(x_{n-1})\big] \\
\; & \approx & \dfrac{h}{2}f(x_0)+hf(x_1)+hf(x_2)+\ldots+hf(x_{n-2})+\dfrac{h}{2}f(x_{n-1})
\end{array}$$ <br>
The cumulative set of weights can be identified by $\displaystyle{\int_a^b}f(x)dx\approx\displaystyle{\sum_{i=0}^{n-1}}c_if(x_i)$. So <br><br>
$$c_i=h\left\{\frac{1}{2},1,1,\ldots 1,\frac{1}{2}\right\}$$

##### <u>Simpson's Rule:</u>
* Let $q=3$ for the integration interpolation: <br><br>
$$\begin{array}{}
L_i(x)=\dfrac{(x-x_{i+1})(x-x_{i+2})}{(\underbrace{x_i-x_{i+1}}_{-h})(\underbrace{x_i-x_{i+2}}_{-2h})} &
L_{i+1}(x)=\dfrac{(x-x_i)(x-x_{i+2})}{(\underbrace{x_{i+1}-x_i}_{h})(\underbrace{x_{i+1}-x_{i+2}}_{-h})} &
L_{i+2}(x)=\dfrac{(x-x_i)(x-x_{i+1})}{(\underbrace{x_{i+2}-x_i}_{2h})(\underbrace{x_{i+2}-x_{i+1}}_{h})}
\end{array}$$ <br>
Let $u=\dfrac{1}{h}(x-x_{i+1})$: <br><br>
$$\begin{array}{}
x=x_i & \Longrightarrow & u=-1 & \dfrac{du}{dx}=\dfrac{1}{h}\\
x=x_{i+2} & \Longrightarrow & u=1 & dx=hdu
\end{array}$$ <br>
* The weights of the interpolating polynomials are: <br><br>
$$\begin{array}{}
w_i & = & \displaystyle{\int_{x_i}^{x_{i+2}}}L_i(x)dx & = & \displaystyle{\frac{h}{2}\int_{-1}^1}u(u-1)du \\
\; & \; & \; & = & \dfrac{h}{2}\displaystyle{\int_{-1}^1}u^2-udu \\
\; & \; & \; & = & \dfrac{h}{2}\left[\dfrac{1}{3}u^3-\dfrac{1}{2}u^2\right]_{-1}^1 \\
\; & \; & \; & = & \dfrac{h}{2}\left[\dfrac{1}{3}(1)^3-\dfrac{1}{3}(-1)^3-\dfrac{1}{2}(1)^2+\dfrac{1}{2}(-1)^2\right] \\
\; & \; & \; & = & \dfrac{h}{2}\left[\dfrac{1}{3}+\dfrac{1}{3}+0\right] \\
\; & \; & \; & = & \dfrac{h}{3}
\end{array}$$ <br>
$$\begin{array}{}
w_{i+1} & = & \displaystyle{\int_{x_i}^{x_{i+2}}}L_{i+1}(x)dx & = & -h\displaystyle{\int_{-1}^1}(u+1)(u-1)du \\
\; & \; & \; & = & -h\displaystyle{\int_{-1}^1}u^2-1du \\
\; & \; & \; & = & -h\left[\dfrac{1}{3}u^3-u\right]_{-1}^1 \\
\; & \; & \; & = & -h\left[\dfrac{1}{3}(1)^3-\dfrac{1}{3}(-1)^3-(1)+(-1)\right] \\
\; & \; & \; & = & -h\left[\dfrac{1}{3}+\dfrac{1}{3}-2\right] \\
\; & \; & \; & = & \dfrac{4h}{3}
\end{array}$$ <br>
$$\begin{array}{}
w_{i+2} & = & \displaystyle{\int_{x_i}^{x_{i+2}}}L_{i+2}(x)dx & = & \displaystyle{\frac{h}{2}\int_{-1}^1}(u+1)udu \\
\; & \; & \; & = & \displaystyle{\frac{h}{2}\int_{-1}^1}u^2+udu \\
\; & \; & \; & = & \dfrac{h}{2}\left[\dfrac{1}{3}u^3+\dfrac{1}{2}u^2\right]_{-1}^1 \\
\; & \; & \; & = & \dfrac{h}{2}\left[\dfrac{1}{3}(1)^3-\dfrac{1}{3}(-1)^3+\dfrac{1}{2}(1)^2-\dfrac{1}{2}(-1)^2\right] \\
\; & \; & \; & = & \dfrac{h}{2}\left[\dfrac{1}{3}+\dfrac{1}{3}+0\right] \\
\; & \; & \; & = & \dfrac{h}{3}
\end{array}$$ <br>
* The composite Simpson's Rule is then: <br><br>
$$\begin{array}{}
\displaystyle{\int_{x_i}^{x_{i+2}}}f(x)dx & \approx & \displaystyle{\sum_{j=0}^{q-1}}w_{i+j}f(x_{i+j}) \; = \; \dfrac{h}{3}\big[f(x_i)+4f(x_{i+1})+f(x_{i+2})\big] \\
\displaystyle{\int_a^b}f(x)dx & = & \displaystyle{\sum_{i=0,2,4,\ldots}^{n-3}\int_{x_i}^{x_{i+2}}}f(x)dx \\
\; & \approx & \dfrac{h}{3}f(x_0)+\dfrac{4h}{3}f(x_1)+\dfrac{2h}{3}f(x_2)+\ldots+\dfrac{2h}{3}f(x_{n-3})+\dfrac{4h}{3}f(x_{n-2})+\dfrac{h}{3}f(x_{n-1})
\end{array}$$ <br>
The cumulative sets of weights are <br><br>
$$c_i=\frac{h}{3}\{1,4,2,4,\ldots,2,4,1\}$$

<u>Errors in Simpsons</u>:
* Taylor expand $f(x)$ around $x_{i+1}$ up to the fourth order, letting $\Delta x\equiv x-x_{i+1}$: <br><br>
$$\begin{array}{}
f(x) & 
= & 
f(x_{i+1})+\Delta xf'(x_{i+1})+\dfrac{1}{2}\Delta x^2f''(x_{i+1})+\dfrac{1}{6}\Delta x^3f'''(x_{i+1})+\dfrac{1}{24}\Delta x^4f^{(4)}(\xi_{i+1})
\end{array}$$ <br>
The integral from $x_i$ to $x_{i+2}$ is then <br><br>
$$\begin{array}{}
\displaystyle{\int_{x_i}^{x_{i+2}}}f(x)dx & 
= & 
\displaystyle{\int_{x_i}^{x_{i+2}}}f(x_{i+1})+\Delta xf'(x_{i+1})+\dfrac{1}{2}\Delta x^2f''(x_{i+1})+\dfrac{1}{6}\Delta x^3f'''(x_{i+1})+\dfrac{1}{24}\Delta x^4f^{(4)}(\xi_{i+1})dx
\end{array}$$ <br>
* Let $u=\dfrac{\Delta x}{h}$: <br><br>
$$\begin{array}{}
x=x_i & \Longrightarrow & u=-1 & \dfrac{du}{dx}=\dfrac{1}{h} \\
x=x_{i+2} & \Longrightarrow & u=1 & du=hdx
\end{array}$$ <br>
$$\begin{array}{}
\displaystyle{\int_{x_i}^{x_{i+2}}}f(x_{i+1})dx &
= &
h\displaystyle{\int_{-1}^1}f(x_{i+1})+huf'(x_{i+1})+\dfrac{1}{2}h^2u^2f''(x_{i+1})+\dfrac{1}{6}h^3u^3f'''(x_i+1)+\dfrac{1}{24}h^4u^4f^{(4)}(\xi_{i+1})du \\
\; &
= &
h\left[uf(x_{i+1})+\dfrac{1}{2}hu^2f'(x_{i+1})+\dfrac{1}{6}h^2u^3f''(x_{i+1})+\dfrac{1}{24}h^3u^4f'''(x_{i+1})+\dfrac{1}{120}h^4u^5f^{(4)}(\xi_{i+1})\right]_{-1}^1 \\
\; &
= &
h\left[2f(x_{i+1})+0+\dfrac{1}{3}h^2f''(x_{i+1})+0+\dfrac{1}{60}h^4f^{(4)}(\xi_{i+1})\right] \\
\; &
= &
2hf(x_{i+1})+\dfrac{1}{3}h^3f''(x_{i+1})+\dfrac{1}{60}h^5f^{(4)}(\xi_{i+1})
\end{array}$$ <br>
from <b><u>Chapter 3: Derivatives</u></b>, the finite difference formula for a second derivative is <br><br>
$$f''(x_{i+1})=\frac{f(x_{i+2})+f(x_{i})-2f(x_{i+1})}{h^2}-\frac{h^2}{12}f^{(4)}(\xi_{i+1})$$ <br>
So our integral becomes <br><br>
$$\begin{array}{}
\displaystyle{\int_{x_i}^{x_{i+2}}}f(x_{i+1})dx &
= &
2hf(x_{i+1})+\dfrac{1}{3}h^3\left(\dfrac{f(x_{i+2})+f(x_i)-2f(x_{i+1})}{h^2}-\dfrac{h^2}{12}f^{(4)}(\xi_{i+1})\right)+\dfrac{1}{60}h^5f^{(4)}(\xi_{i+1}) \\
\; &
= &
2hf(x_{i+1})+\dfrac{1}{3}hf(x_{i+2})+\dfrac{1}{3}hf(x_i)-\dfrac{2}{3}hf(x_{i+1})-\dfrac{h^5}{36}f^{(4)}(\xi_{i+1})+\dfrac{1}{60}h^5f^{(4)}(\xi_{i+1}) \\
\; &
= &
\dfrac{h}{3}\big[f(x_i)+4f(x_{i+1})+f(x_{i+2})\big]-\dfrac{1}{90}h^5f^{(4)}(\xi_{i+1})
\end{array}$$ <br>
Comparing to Simpson's Rule, $\displaystyle{\int_{x_i}^{x_{i+2}}}f(x)dx\approx\dfrac{h}{3}\big[f(x_i)+4f(x_{i+1})+f(x_{i+2})\big]$, the absolute error is then <br><br>
$$\varepsilon_i=-\frac{1}{90}h^5f^{(4)}(\xi_{i+1})$$ <br>
From the sum of the composite Simpson's rule, the total error of the approximation becomes <br><br>
$$\varepsilon=\sum_{i=0,2,4,\ldots}^{n-3}\varepsilon_i\Longrightarrow\varepsilon=-\dfrac{1}{90}h^5\sum_{i=0,2,4,...}^{n-3}f^{(4)}(\xi_{i+1})$$ <br>
$$\varepsilon=-\dfrac{1}{90}h^5\big(f^{(4)}(\xi_1)+f^{(4)}(\xi_3)+f^{(4)}(\xi_5)+\ldots+f^{(4)}(\xi_{n-2})\big)$$ <br>
and, again, $\exists$ $\xi$ s.t. $f^{(4)}(\xi)=\bar{f}^{(4)}(\xi_{2n-1})$ for $n\in\mathbb{N}$ (i.e., Mean Value Theorem); introduce $\dfrac{(n-1)/2}{(n-1)/2}$ as the coefficient (since the absolute error per abscissas are the odd, positive integers) to yield <br><br>
$$\begin{array}{}
\varepsilon &
= &
-\dfrac{(n-1)/2}{90}h^5\dfrac{f^{(4)}(\xi_1)+f^{(4)}(\xi_3)+f^{(4)}(\xi_5)+\ldots+f^{(4)}(\xi_{n-2})}{(n-1)/2} \\
\; &
= &
-\dfrac{n-1}{180}h^5f^{(4)}(\xi) \\
\; &
= &
-\dfrac{b-a}{180}h^4f^{(4)}(\xi) \;^*
\end{array}$$ <br>
<sup>* Recall: $h=\frac{b-a}{n-1}$</sup> <br><br>
Our total error in $\underline{\underline{\text{Simpson's 1/3 Rule}}}$ ($\text{1/3}$ for the quadratic interpolation used for approximation) is finally $\varepsilon=-\dfrac{b-a}{180}h^4f^{(4)}(\xi)$. <br><br>
    - For polynomials up to the third degree, Simpson's (1/3) Rule gives exact answers. <br><br>
    - Most typical approxiation methods for integration.

### Comparing Left-hand Rectangle Rule, Trapezoid Rule, and Simpson's (1/3) Rule as open methods for approximating the integral of $\frac{1}{\sqrt{x^2+1}}$.

```python
import matplotlib.pyplot as plt

def trapezoid(f, xs):
    a, b, n, h = params(xs)
    ws = np.array([0.5 if i == 0 or i == n-1 else 1 for i in range(0,n)])
    trap = h*(ws*f(xs)).sum()
    return trap

def simpson(f, xs):
    a, b, n, h = params(xs)
    ws = 2*np.ones(n)
    ws[1::2] = 4
    ws[0] = 1.
    ws[-1] = 1.
    simp = (h/3)*(ws*f(xs)).sum()
    return simp

xs = np.linspace(0.,1.,51)
print(f"analytical = {anal_int}")
print(f"rectangle = {lefthand(f, xs)}")
print(f"trapezoid = {trapezoid(f, xs)}")
print(f"simpson = {simpson(f, xs)}")

#simpson's method needs an even number of data points
ns = np.logspace(1,7,20).astype(int)
ns[ns%2 == 0] += 1

rect_ints = [lefthand(f, np.linspace(0.,1.,n)) for n in ns]
trap_ints = [trapezoid(f, np.linspace(0.,1.,n)) for n in ns]
simp_ints = [simpson(f, np.linspace(0.,1.,n)) for n in ns]

plt.plot(ns, abs(rect_ints-anal_int), "rs", label="Left-hand Rectangle Rule")
plt.plot(ns, abs(trap_ints-anal_int), "bd", label="Trapezoid Rule")
plt.plot(ns, abs(simp_ints-anal_int), "mx", label="Simpson's Rule")
plt.xscale("log")
plt.xlabel(r"$n$")
plt.yscale("log")
plt.ylabel(r"$|\varepsilon|$")
plt.title("Total absolute errors of Rectangle, Trapezoid, and Simpson's Rules")
plt.legend()
```

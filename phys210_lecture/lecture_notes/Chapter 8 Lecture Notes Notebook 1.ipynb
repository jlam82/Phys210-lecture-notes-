# <u>Chapter 8: Differential Equation</u>

$\underline{\mathcal{Ex:}}$ Projectile motion in two-dimensions with drag:
> $$\begin{array}{}
\vec{F}=m\vec{a} & \vec{a}=\dfrac{d^2\vec{r}}{dt^2} & \vec{r}=x\hat{x}+y\hat{y}
\end{array}$$ <br>
Drag force (for large enough $|\vec{v}|$): $\vec{F_d}=-kmv^2\hat{v}$ <br><br>
$$\begin{array}{}
\dfrac{d^2x}{dt^2}=-k\dfrac{dx}{dt}\sqrt{\left(\dfrac{dx}{dt}\right)^2+\left(\dfrac{dy}{dt}\right)^2} & 
\; &
\dfrac{d^2y}{dt^2}=-g-k\dfrac{dy}{dt}\sqrt{\left(\dfrac{dx}{dt}\right)^2+\left(\dfrac{dy}{dt}\right)^2}
\end{array}$$ <br>
Two simultaneous ordinary differential equations for $x(t)$ and $y(t)$. <br><br>
$\bullet$ If given $x(0)$, $y(0)$, $v_x(0)$, $v_y(0)$, we have an Initial Value Problem (IVP). <br><br>
$\bullet$ If given $x(0)$, $y(0)$, $v_x(0)$, $y(\tau)=0$, we have a Boundary Value Problem (BVP).
>

$\underline{\mathcal{Ex:}}$ <code>Schr&ouml;dinger Equation(s)</code>
> Single spinless non-relativistic particle of mass $m$ in one spatial dimension $x$ described by its <i>wave function</i> $\Psi(x,t)$ that satisfies the <i>time-dependent Schrodinger equation</i>: <br><br>
$$i\hbar\frac{\partial\Psi(x,t)}{\partial t}=-\frac{\hbar^2}{2m}\frac{\partial^2\Psi(x,t)}{\partial x^2}+V(x,t)\Psi(x,t)$$ <br>
where $-\frac{\hbar^2}{2m}\frac{\partial^2\Psi(x,t)}{\partial x^2}$ is the <i>kinetic</i> energy term while $V(x,t)\Psi(x,t)$ is the <i>potential</i> energy term. <br><br>
$\bullet$ A linear partial differential equation IVP for determining how $\Psi$ changes in time as a dynamical equation. <br><br>
If the potential energy term is time-independent, i.e., $V(x)$, then the time-dependent Schrodinger equation is <i>seperable</i> via <br><br>
$$\Psi(x,t)=\psi(x)e^{-i\tfrac{E}{\hbar}t}$$ <br>
where $\psi(x)$ satisfies the <i>time-independent Schrodinger equation</i>: <br><br>
$$E\psi(x)=-\frac{\hbar^2}{2m}\frac{d^2\psi(x)}{dx^2}+V(x)\psi(x)$$ <br>
$\bullet$ The time-independent Schrodinger equation is now an ordinary differential equation EVP <sup>*</sup>, where $E$ is a real number and the equation is no longer dynamical. <br><br>
<sup>* : EigenValue Problem (EVP)</sup>
>

##### <u>Ordinary Differential Equations (ODE):</u>
We start with a single-variable differential equation for $y(x)$, where
* $x$ is our independent variable. <br><br>
* $y$ is our dependent variable.

$$\begin{array}{}
y'(x)=f\big(x,y(x)\big) &
\; &
y(a)=c
\end{array}$$ <br>
where $y'=\frac{dy}{dx}$ and $f(x,y)$ is known. <br><br>
We want to solve for $y$: <br><br>
$$\int \dfrac{dy}{dx}=\int f\big(x', y(x')\big)dx'$$ <br>
$$y(x)=c+\int_a^xf\big(z,y(z)\big)dz$$ <br>
Note that if $f\equiv f(z)$ and not $y$, then we need to find $\displaystyle{\int}f(z)dz$ <sup>*</sup> <br><br>
<sup>* : via Trapezoidal Approximation and Simpson's Rule from <b>Chapter 7: Integrals</b>.</sup> <br><br>
We may also have higher-order ODE's:
* Second-order:
$$\begin{array}{}
y''=f(x,y,y') & \underbrace{y(a)=c \;\;\;\;\; y'(a)=d}_{\text{IVP}}
\end{array}$$
Or
$$\begin{array}{}
y''=f(x,y,y') & \underbrace{y(a)=c \;\;\;\;\; y(b)=d}_{\text{BVP}}
\end{array}$$
Or
$$\begin{array}{}
y''=f(x,y,y';s) & \underbrace{y(a)=c \;\;\;\;\; y(b)=d}_{\text{EVP}}
\end{array}$$
where $s$ is a parameter; EVP only has non-trivial solutions for some values of $s$.

<u>IVP's:</u> <br><br>
Solve for $y(x)$ with $x\in[a,b]$. <br><br>
$$\begin{array}{}
y'(x)=\underbrace{f\big(x,y(x)\big)}_{\text{known}} & 
y(a)=c
\end{array}$$ <br>
We will typically solve for $y(x)$ at $n$ grid points $x_j$: <br><br>
$$\begin{array}{}
x_j=a+jh & j=0,1,\ldots,n-1
\end{array}$$ <br>
and $h=\dfrac{b-a}{n-1}$. <br><br>
Notations: $y(x_j)$ is an exact ODE solution at $x_j$ while $y_j$ is an approximate solution.
* <u>Forward Euler's Method:</u> <br><br>
Recall the formula for calculating the forward difference of a function: <br><br>
$$f'(x)=\frac{f(x+h)-f(x)}{h}-\frac{h}{2}f''(x)+\ldots$$ <br>
Our ODE solution is then <br><br>
$$y'(x_j)=\frac{y(x_{j+1})-y(x_j)}{h}-\frac{h}{2}y''(\xi_j)$$ <br>
where $\xi_j$ is a point between $x_j$ and $x_{j+1}$ and $y(x_j)$ is an exact solution to our ODE. <br><br>
By definition, $y'(x_j)=f\big(x_j,y(x_j)\big)$. Rearranging, <br><br>
$$y(x_{j+1})=y(x_j)+hf\big(x_j,y(x_j)\big)+\frac{h^2}{2}y''(xi_j)$$ <br>
Using $y_j$ is our substitution, our approximation is then <br><br>
$$\begin{array}{}
y_{j+1}=y_j+hf(x_j,y_j) & j=0,1,\ldots,n-2 & y_0=c
\end{array}$$ <br>
where $f(x_j, y_j)$ is an approximation to the slope of the tangent to the exact solution at $x_j$. <br><br>
The total error $\varepsilon$ that accumulates from $x_0=a$ to $x_{n-1}=b$ is <br><br>
$$|\varepsilon|=|y(b)-y_{n-1}|\leq h\times\text{const.}$$ <br>
Overall, the forward Euler's method converges and its total error decreases linearly, $\mathcal{O}(h)$, i.e., $\varepsilon\rightarrow 0$ as $h\rightarrow 0$. <br><br>
    - Stability: examine a "test equation" <br><br>
    $$\begin{array}{}
    y'(x)=\mu y(x) & y(0)=1 & \mu\in\mathbb{R}
    \end{array}$$ <br>
    where the exact solution is $y(x)=e^{\mu x}$. <br><br>
    Our approximation is then <br><br>
    $$y_{j+1}=y_j+h\mu y_j=y_j(1+h\mu)$$ <br>
    $$\left\{
    \begin{array}{l}
    y_1=y_0(1+h\mu) \\
    y_2=y_1(1+h\mu)=y_0(1+h\mu)^2 \\
    y_3=y_2(1+h\mu)=y_0(1+h\mu)^3 \\
    \;\;\;\;\; \;\;\;\;\; \;\;\;\;\; \;\;\;\;\; \;\;\; \vdots \\
    y_{n-1}=y_0(1+h\mu)^{n-1}=(1+h\mu)^{n-1}\text{ since }y_0=1
    \end{array}
    \right.$$ <br>
    Consider when $\mu<0$ (i.e., exponential decay). In order for $|y_{j+1}|<|y_j|$, we need <br><br>
    $$|1+h\mu|<1\Longrightarrow h<\frac{2}{|\mu|}$$

### Forward Euler Method and Stability for $y'(x)=-20y(x)$

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x,y: -20.*y

def fw_euler(f, a, b, n, y0):
    h = (b-a)/(n-1)
    xs = a+np.arange(n)*h
    ys = np.zeros(n)
    y = y0
    for i,x in enumerate(xs):
        ys[i] = y
        y += h*f(x,y)
    return xs, ys

xs = np.linspace(0, 1, 50)
plt.plot(xs, np.exp(-20*xs), "k-", label="exact")

y0 = 1

n1 = 10
xs1, ys1 = fw_euler(f, 0, 1, n1, y0)
plt.plot(xs1, ys1, "ro--", label="$n=10$")

n2 = 12
xs2, ys2 = fw_euler(f, 0, 1, n2, y0)
plt.plot(xs2, ys2, "bd--", label="$n=12$")

plt.legend()
```

* <u>Backward Euler's Method:</u> <br><br>
Analagous to forward Euler's method except with the backward difference approximation: <br><br>
$$\begin{array}{}
\underline{\text{Formula:}} & \underbrace{y'(x_{j+1})}_{f\left(x_{j+1},\;y(x_{j+1})\right)}=\dfrac{y(x_{j+1})-y(x_j)}{h}+\dfrac{h}{2}y''(\xi_j) \\
\underline{\text{Rearrangement:}} & y(x_{j+1})=y(x_j)+hf\big(x_{j+1}, y(x_{j+1})\big)-\dfrac{h^2}{2}y''(\xi_j) \\
\; \\
\underline{\text{Approximation:}} & y_{j+1}=y_j+hf(x_{j+1}, y_{j+1}) \;\;\;\;\; j=0,1,\ldots,n-2 \;\;\;\;\; y_0=c
\end{array}$$ <br>
Looks very similar, but actually a big difference: <br><br>
$$\begin{array}{}
\underline{\underline{y_{j+1}}}=y_j+hf(x_{j+1}, \underline{\underline{y_{j+1}}}) & _{*:\underline{\underline{\text{ Need } y_{j+1}\text{ to find }y_{j+1}}}} \\
\; \\
z=y_j+hf(x_{j+1},z) & _{\underline{\underline{\text{Implicit}}}}
\end{array}$$ <br>
Consider $y'(x)=\mu y(x)$ (so the exact solution is $y(x)=e^{\mu x}$) <br><br>
$$y_{j+1}=y_j+h\mu y_{j+1}\Longrightarrow y_{j+1}=\frac{y_j}{1-\mu h}$$ <br>
when $\mu<0$, we always get $|y_{j+1}|<|y_j|$, unconditionally stable solution.

### Backward Euler Method and Stability for $y'(x)=-20y(x)$

```python
def bw_euler(f, a, b, n, y0, mu):
    h = (b-a)/(n-1)
    xs = a + np.arange(n)*h
    ys = np.zeros(n)
    y = y0
    for i,x in enumerate(xs):
        ys[i] = y
        y *= 1./(1-mu*h)
    return xs, ys

y0 = 1
mu = -20

xs = np.linspace(0, 1, 50)
plt.plot(xs, np.exp(-20*xs), "k-", label="exact")

n1 = 12
xs1, ys1 = fw_euler(f, 0, 1, n1, y0)
plt.plot(xs1, ys1, "bd--", label="Forward Euler $n=12$")

n2 = 10
xs2, ys2 = bw_euler(f, 0, 1, n2, y0, mu)
plt.plot(xs2, ys2, "gs--", label="Backward Euler $n=10$")

plt.legend()
```

<u>2<sup>nd</sup> order Runge-Kutta Methods:</u> <br><br>
Start with a Taylor expansion to the 4<sup>th</sup> order:<br><br>
$$\begin{array}{}
    y(x_{j+1}) & = & y(x_j)+hy'(x_j)+\frac{h^2}{2}y''(x_j)+\frac{h^3}{6}y'''(x_j)+\mathcal{O}(h^4) \\
    \; & = & y(x_j)+hf\big(x_j,y(x_j)\big)+\dfrac{h^2}{2}f'\big(x_j,y(x_j)\big)+\dfrac{h^3}
    {6}f''\big(x_j,y(x_j)\big)+\mathcal{O}(h^4) \\
    \; & = & y(x_j)+hf\big(x_j,y(x_j)\big)+\dfrac{h^2}{2}\left(\dfrac{\partial f}{\partial x}+f\dfrac{\partial f}{\partial 
    y}\right)^{*_1}+\dfrac{h^3}{6}\left(\dfrac{\partial^2 f}{\partial x^2}+2f\dfrac{\partial^2 f}{\partial x\partial 
    y}+\dfrac{\partial f}{\partial x}\dfrac{\partial f}{\partial y}+f^2\dfrac{\partial^2f}{\partial 
    y^2}+f\left(\dfrac{\partial f}{\partial y}\right)^2\right)^{*_2}+\mathcal{O}(h^4)
\end{array}$$
<br>
$^{ %^ can be added or deleted to control size of the notes
    \begin{gather}
        {\begin{array}{l}
            *_1 : & f'=\frac{df}{dx}\\
            \; & \frac{df}{dx}=\frac{\partial f}{\partial x}+\frac{\partial f}{\partial y}\frac{dy}{dx} \\
            \; & f\overset{\mathrm{def}}{\Longleftrightarrow}\frac{dy}{dx} \\
            \; & f'=\frac{\partial f}{\partial x}+\frac{\partial f}{\partial y}f \\
        \end{array}}
    &
        {\begin{array}{l}
            *_2: & f''=\frac{d^2f}{dx^2} \; \Longrightarrow \; \frac{d^2f}{dx^2}=\frac{d}{dx}\left(\frac{df}{dx}\right) \; 
            \Longrightarrow \; \frac{d}{dx}\left(f'\right) \\ 
            \; & \frac{d^2f}{dx^2}=\frac{\partial f'}{\partial x}+\frac{\partial f'}{\partial y}\frac{dy}{dx} \\ 
            \; & \frac{d^2f}{dx^2}=\frac{\partial}{\partial x}\left(\frac{\partial f}{\partial x}+\frac{\partial f}
            {\partial y}f\right)+\frac{\partial}{\partial y}\left(\frac{\partial f}{\partial x}+\frac{\partial f}{\partial 
            y}f\right)f \\ 
            \; & \frac{d^2f}{dx^2}=\frac{\partial^2f}{\partial x^2}+\left(\frac{\partial^2f}{\partial x\partial 
            y}f+\frac{\partial f}{\partial y}\frac{\partial f}{\partial x}\right)+\frac{\partial^2f}{\partial y\partial 
            x}f+\left(\frac{\partial^2f}{\partial y^2}f^2+\frac{\partial f}{\partial y}\frac{\partial f}{\partial 
            y}f\right) \; ^{**_2\text{ : Product rule, then distribute }f} \\ 
            \; & \frac{d^2f}{dx^2}=\frac{\partial^2f}{\partial x^2}+2\frac{\partial^2f}{\partial x\partial 
            y}f+\frac{\partial f}{\partial x}\frac{\partial f}{\partial y}+\frac{\partial^2f}{\partial 
            y^2}f^2+\left(\frac{\partial f}{\partial y}\right)^2f \;^{**_2\text{: Clairaut's Theorem for}\frac{\partial^2f}
            {\partial x\partial y}=\frac{\partial^2f}{\partial y\partial x}}
        \end{array}}
    \end{gather}
}$

Match the Taylor expansion to a <i>second-order Runge-Kutta</i> approximation: <br><br>
$$y_{j+1}=y_j+c_0hf(x_j,y_j)+c_1hf\left(x_j+c_2h,\;y_j+c_2hf(x_j,y_j)\right)$$ <br>
Need to determine $c_0$, $c_1$, $c_2$. <br><br>
Let $k_0=hf(x_j,y_j)$: <br><br>
$$y_{j+1}=y_j+c_0k_0+c_1hf(x_j+c_2h,\;y_j+c_2k_0)$$ <br>
It may be seen then that the right-hand side of the equation only requires two function evaluation. If $c_0$, $c_1$, $c_2$, $x_j$, $y_j$, and $f$ are known, then we can evaluate it <i>explicitly</i>. <br><br>
Taylor expand the last term: <br><br>
$$f\left(x_j+c_2h,\;y_j+c_2hf(x_j,y_j)\right)=f(x_j, y_j)+c_2h\frac{\partial f}{\partial x}+c_2hf(x_j,y_j)\frac{\partial f}{\partial y}+\frac{c_2^2h^2}{2}\frac{\partial^2f}{\partial x^2}+\frac{c_2^2h^2}{2}2f(x_j,y_j)\frac{\partial^2f}{\partial x \partial y}+\frac{c_2^2h^2}{2}f^2(x_j,y_j)\frac{\partial^2f}{\partial y^2}+\mathcal{O}(h^3)$$ <br>
and substitute back to our approximation before factoring out like terms to get <br><br>
$$y_{j+1}=y_j+(c_0+c_1)hf(x_j,y_j)+c_1c_2h^2\left(\frac{\partial f}{\partial x}+f(x_j,y_j)\frac{\partial f}{\partial y}\right)+c_1c_2^2\frac{h^3}{2}\left(\frac{\partial^2f}{\partial x^2}+2f(x_j,y_j)\frac{\partial^2f}{\partial x\partial y}+f^2(x_j,y_j)\frac{\partial^2f}{\partial y^2}\right)+\mathcal{O}(h^4)$$ <br>
If we compare our approximation to the 4<sup>th</sup> order Taylor expansion of our exact solution, we can match the two for the $h$ and $h^2$ terms if we let <br><br>
$$\begin{array}{}
    c_0+c_1=1 & c_1c_2=\dfrac{1}{2}
\end{array}$$ <br>
However, we have no way to match our $h^3$ term since our last expression is missing the terms $\frac{\partial f}{\partial x}\frac{\partial f}{\partial y}$ and $f\left(\frac{\partial^2f}{\partial y^2}\right)^2$. Regardless, we then have our <i>2<sup>nd</sup> order Runge-Kutta</i> equation up to an error of $\mathcal{O}(h^3)$.
* <u>Explicit Midpoint Method:</u> <br><br>
There are various ways to determine $c_0$, $c_1$, and $c_2$ when matching yielded $c_0+c_1=1$ and $c_1c_2=\frac{1}{2}$. One solution is the following: <br><br>
$$\begin{array}{}
    c_0=0 & c_1=1 & c_2=\dfrac{1}{2}
\end{array}$$ <br>
the 2<sup>nd</sup> order Runge-Kutta equation then becomes <br><br>
$$y_{j+1}=y_j+hf\left(x_j+\tfrac{h}{2},\;y_j+\tfrac{h}{2}f(x_j,y_j)\right)$$ <br>
This is referred to as an explicit method because the right-hand side of the equation is already known, and it is a midpoint method because we find a midpoint between $x_j$ and $x_{j+1}$. Specifically, we find an approximate slope of the tangent at this midpoint and shift it to $(x_j,y_j)$ to determine $y_{j+1}$. <br><br>
* <u>Explicit Trapezoid Method:</u> <br><br>
Another solution to determining $c_0$, $c_1$, and $c_2$ is <br><br>
$$\begin{array}{}
    c_0=\dfrac{1}{2} & c_1=\dfrac{1}{2} & c_1=1
\end{array}$$ <br>
the 2<sup>nd</sup> order Runge-Kutta equation again then becomes <br><br>
$$y_{j+1}=y_j+\tfrac{h}{2}f(x_j,y_j)+\tfrac{h}{2}f\left(x_{j+1},\;y_j+hf(x_j, y_j)\right)$$ <br>
$$y_{j+1}=y_j+\tfrac{h}{2}\Big(f(x_j,y_j)+f\big(x_{j+1},\;y_j+hf(x_j,y_j)\big)\Big)$$ <br>
Likewise, this is an explicit method, and it is a trapezoid method for the $\tfrac{h}{2}$ term being multiplied to the sum of (what appears to be) function evaluations of a left endpoint and right endpoint, $x_j$ and $x_{j+1}$.

Recall that our goal is to determine $y_{j+1}$, an approximation to $y(x_{j+1})$, which is our exact solution to our ODE. Determining either of these would require integration, and more specifically integral approximation for ODE's that can't be solved analytically. Knowing this, we have: <br><br>
$$y(x_{j+1})-y(x_j)=\int_{x_j}^{x_{j+1}}\frac{dy}{dx}dx=\int_{x_j}^{x_{j+1}}f\big(x,y(x)\big)dx$$ <br>
But our integral of $f$ depends on both $x$ and $y$, which in turn also depends on $x$. <br><br>
We may approximate our integral via the following: <br><br>
$\;\;\;\;\;
\begin{array}{}
    \bullet \; \text{Left-hand Rectangle:} & \bullet \; \text{Right-hand Rectangle:} & \bullet \; \text{Midpoint Rule:} &
    \bullet \; \text{Trapezoid Rule:}\\
    y_{j+1}=\underbrace{y_j+hf(x_j,y_j)}_{\text{Forward Euler, Explicit}} & 
    y_{j+1}=\underbrace{y_j+hf(x_{j+1},y_{j+1})}_{\text{Backward Euler, Implicit}} & 
    y_{j+1}=y_j+hf\big(x_j+\frac{h}{2},\;\underbrace{y\left(x_j+\frac{h}{2}\right)}_{\text{Unknown}}\big) & 
    y_{j+1}=\underbrace{y_j+\frac{h}{2}\left(f(x_j, y_j)+f(x_{j+1}, y_{j+1})\right)}_{\text{Implicit Trapezoid Method}} \\
    \; & \; & \underset{\text{Forward Euler, Explicit Midpoint}}{y\left(x_j+\frac{h}{2}\right)\approx\frac{1}{2}
    (y_j+\underbrace{y_{j+1}})} & 
    \underbrace{y_{j+1}}_{\text{RHS}}=\underbrace{y_j+hf(x_j,y_j)}_{\text{Forward Euler Method}} \\
    \; & \; & y_{j+1}=\underbrace{y_j+hf\left(x_j+\frac{h}{2}, \frac{y_j+y_{j+1}}{2}\right)}_{\text{Implicit 
    Midpoint Method}} & 
    y_{j+1}=\underbrace{y_j+\frac{h}{2}\Big(f(x_j,y_j)+f\big(x_{j+1},\;y_j+hf(x_j,y_j)\big)\Big)}_{\text{Explicit Trapezoid Method}}\\
\end{array}$

### 2<sup>nd</sup> Order Runge-Kutta Midpoint Rule for $y'(x)=-20y(x)$

```python
def RK2_mid(f, a, b, n, y0):
    h = (b-a)/(n-1)
    xs = a + np.arange(n)*h
    ys = np.zeros(n)
    y = y0
    for i,x in enumerate(xs):
        ys[i] = y
        k0 = h*f(x,y)
        y += h*f(x+h/2, y+k0/2)
    return xs, ys

y0 = 1

xs = np.linspace(0, 1, 50)
plt.plot(xs, np.exp(-20*xs), "k-", label="exact")

n1 = 12
xs1, ys1 = RK2_mid(f, 0, 1, n1, y0)
plt.plot(xs1, ys1, "gs--", label="RK2 Midpoint Rule for $n=12$")

n2 = 24
xs2, ys2 = RK2_mid(f, 0, 1, n2, y0)
plt.plot(xs2, ys2, "ro--", label="RK2 Midpoint Rule for $n=24$")

plt.legend()
```

<u>4<sup>th</sup> order Runge-Kutta Method:</u> <br><br>
Returning back to the Taylor expansion of $y(x_{j+1})$ to the 4<sup>th</sup> order, a very important Ansatz<sup>*</sup> must be presented: <br><br>
$$\begin{align}
    k_0 & = hf(x_j, y_j) \\
    k_1 & = hf(x_j+\tfrac{h}{2},\; y_j+\tfrac{k_0}{2}) \\
    k_2 & = hf(x_j+\tfrac{h}{2},\; y_j+\tfrac{k_1}{2}) \\
    k_3 & = hf(x_j+h,\; y_j+k_2) \\
\end{align}$$ <br>
$$y_{j+1}= y_j+\tfrac{1}{6}(k_0+2k_1+2k_2+k_3)+\mathcal{O}(h^5)$$ <br>
<sup>*: an educated guess or additional assumption made to help solve a problem, and which may later be verified to be part of the solution by its results.</sup> <br>
typically known as <i>classic Runge-Kutta</i> or <i>RK4</i>. Like its predecessor, RK4 does not need to evaluate any derivatives for $y_{j+1}$ $-$ just four function evaulations instead. At each evaluation $k_i$ for $i\in\{0,1,2,3\}$, an approximation to the slope between an endpoint and midpoint is made.

```python
import numpy as np #little different from what in the textbook for personal readability 

def f(x,y):
    return - (30/(1-x**2)) + ((2*x)/(1-x**2))*y - y**2

def euler(f, a, b, n, y0):
    h = (b-a)/(n-1)
    xs = a + np.arange(n)*h
    ys = np.zeros(n)
    y = y0
    for j,x in enumerate(xs):
        ys[j] = y
        y += h*f(x,y)
    return xs, ys

def rk4(f, a, b, n, y0):
    h = (b-a)/(n-1)
    xs = a + np.arange(n)*h
    ys = np.zeros(n)
    y = y0
    for j,x in enumerate(xs):
        ys[j] = y
        k0 = h*f(x,y)
        k1 = h*f(x+h/2, y+k0/2)
        k2 = h*f(x+h/2, y+k1/2)
        k3 = h*f(x+h, x+k2)
        y += (k0 + 2*k1 + 2*k2 + k3)/6
    return xs, ys

if __name__ == "__main__":
    a, b, n, y0 = 0.05, 0.49, 12, 19.53
    xs, ys = euler(f, a, b, n, y0); print(ys)
    xs, ys = rk4(f, a, b, n, y0); print(ys)
```

# <u>Chapter 2: Errors</u>

### <u>Experimental Errors</u>
<b>Experimental Errors</b> are random and systematic.
* <b>Random:</b> not repeatable, lead to fluctuations due to <br><br>

    - precision of measuring equipment (e.g., try to measure the width of a 
      human hair with a meter stick). <br><br>
      
    - fluctuations in environment (temporary). <br><br>
    
    - true randomness (e.g., radioactive decay). <br><br>

* <b>Systematic</b>: repeatable and based on how the experimenter chooses to measure. <br><br>

    - Operator Error. <br><br>
    
    - Uncalibrated measuring device. <br><br>
    
    - Wrong theory or approximation. (e.g., small angle approximation for pendulum)

### <u>Accuracy and Precision</u>

* <b>Precision</b> is limited by random errors. <br><br>

    - Closeness of grouping <br><br>
    
    - May or may not get "correct" value <br><br>
    
    - Estimate by repetition (i.e., standard deviation). <br><br>

```python
import numpy as np
import matplotlib.pyplot as plt

xs = np.linspace(-0.8, 0.8, 100)
f = lambda x: 1/np.sqrt(1-x)
f0 = lambda x: np.ones(x.size)
f1 = lambda x: np.ones(x.size) + x/2
f2 = lambda x: np.ones(x.size) + x/2 + 3*x**2/8
f3 = lambda x: np.ones(x.size) + x/2 + 3*x**2/8 + 5*x**3/16

plt.plot(xs, f(xs), "k-", label=r"$\frac{1}{\sqrt{1-x}}$")
plt.plot(xs, f0(xs), "r--", label=r"$1$")
plt.plot(xs, f1(xs), "b--", label=r"$1+\frac{1}{2}x$")
plt.plot(xs, f2(xs), "g--", label=r"$1+\frac{1}{2}x+\frac{3}{8}x^2$")
plt.plot(xs, f3(xs), "m--", label=r"$1+\frac{1}{2}x+\frac{3}{8}x^2+\frac{5}{16}x^3$")

plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.ylim(-0.1, 2.5)
plt.grid(linestyle="--")
plt.legend()

plt.show()
```

### <u>Uncertainty Propogation:</u>
* We measure $t$ with some uncertainty $\delta t$ and want to find the uncertainty in $f(t)$ (i.e., $f(t)=\frac{2y}{t^2}$).

$\delta f\cong f^{\prime}(t_0)\delta t$

<ul>
    <li>
        For many variables $f(t,r,s...)$,
        $\newcommand{\parder}[2][]{ \dfrac{\partial #1}{\partial #2} }$
        $$\delta f^2\approx\left(\parder[f]{t}\right)^2\delta t^2 + \left(\parder[f]{r}\right)^2\delta r^2 + \left(
        \parder[f]{s}\right)^2\delta s^2 + \ldots$$
    </li> <br>
    $\underline{\mathcal{Ex}:}$ for the uncertainty in an area $A$ given by the formula $A=lw$,
    <table style="width:100%" align="left">
        <tr>
            <td style="text-align:left">1. Measure $l_0\pm\delta l$ and $w_0\pm\delta w$. Let $A=l_0w_0\pm\delta A$</td>
        </tr>
        <tr>
            <td style="text-align:left">2. Use the definition of $\delta f^2$ to express $\delta A^2=\left(\parder[A]{l}
            \right)^2\delta l^2 + \left(\parder[A]{w}\right)^2\delta w^2$. </td>
        </tr>
        <tr>
            <td style="text-align:left">3. The partial derivatives simplify to $\delta A^2=w^2\delta l^2 + l^2\delta w^2$.
            </td>
        </tr>
        <tr>
            <td style="text-align:left">4. To help express a definition later on, divide both sides with $A^2$. 
            So $\dfrac{\delta A^2}{A^2}=\dfrac{w^2\delta l^2}{A^2} + \dfrac{l^2\delta w^2}{A^2}$</td>
        </tr>
        <tr>
            <td style="text-align:left">5. Express the $A^2$ in the denominator with its definition, $A=lw$: 
    </table>
</ul>

<ul>
    <li>
        For many variables $f(t,r,s...)$,
        $\newcommand{\parder}[2][]{ \dfrac{\partial #1}{\partial #2} }$
        $$\delta f^2\approx\left(\parder[f]{t}\right)^2\delta t^2 + \left(\parder[f]{r}\right)^2\delta r^2 + \left(
        \parder[f]{s}\right)^2\delta s^2 + \ldots$$
    </li> <br>
    $\underline{\mathcal{Ex}:}$ for the uncertainty in an area $A$ given by the formula $A=lw$,
    <ol>
        <li>
            Measure $l_0\pm \delta l$ and $w_0\pm \delta w$
        </li> <br>
        <li>
            So $A=l_0w_0\pm \delta A$
        </li> <br>
        <li>
            Use the definition of $\delta f^2$ to express $\delta A^2=\left(\frac{\partial A}{\partial l}\right)^2\delta 
            l^2+\left(\frac{\partial A}{\partial w}\right)^2\delta w^2$
    </ol>
</ul>

>  <br><br>
  $\delta A^2=w^2\delta l^2+l^2\delta w^2$ <br><br>
  $\frac{\delta A^2}{A^2}=\frac{w^2\delta l^2}{A^2}+\frac{l^2\delta w^2}
  {A^2}$ <br><br>
  $(\frac{\delta A}{A})^2=\frac{w^2\delta l^2}{l^2w^2}+\frac{l^2\delta w^2}
  {l^2w^2}$ <br><br>
  $(\frac{\delta A}{A})^2=\frac{\delta l^2}{l^2}+\frac{\delta w^2}{w^2}$ 
  <br><br>
  $(\frac{\delta A}{A})^2=(\frac{\delta l}{l})^2+(\frac{\delta w}{w})^2$
>
    Let's say $l_0=100\pm1mm$ and $w_0=10\pm1mm$, therefore $A_0=1000mm^2$
> $(\frac{\delta A}{A})^2=(1mm/100mm)^2+(1mm/10mm)^2$ <br><br>
  $(\frac{\delta A}{A})^2=(1\%)^2+(10\%)^2$ <br><br>
  $(\frac{\delta A}{A})^2\cong(10\%)^2$ <br><br>
  $\frac{\delta A}{|A|}\cong10\%$
>
    And such a new definition called <b>fractional uncertainty</b> is shown: <br><br>
    - <b>Fractional Uncertainty</b>: the absolute uncertainty divided by 
    the quantity itself (note that the units cancel in the division, so 
    fractional uncertainty are unitless) <br><br>
* For our gravity lab, what is our uncertainty in $\bar{t}$? Is it sample standard deviation $S(t_i)$?
> $\bar{t}=\frac{1}{5}[t_1+t_2+...+t_5]$ <br><br>
$(\delta\bar{t})^2=(\frac{\partial\bar{t}}{\partial t_1})^2\delta t_1^2+(\frac{\partial\bar{t}}{\partial t_2})^2\delta t_2^2+...+(\frac{\partial\bar{t}}{\partial t_5})^2\delta t_5^2$ <br><br>
<u>Assume</u>: $\delta t_1=\delta t_2=...=\delta t_5=\delta t$ <br><br>
We estimate $\delta t$ by $S=\big[\frac{1}{N-1}\Sigma(t_i-\bar{t})^2\big]^{1/2}$ <br><br>
$(\delta\bar{t})^2=\big(\frac{1}{5}\big)^2S^2+\big(\frac{1}{5}\big)^2S^2...+\big(\frac{1}{5}\big)^2S^2$ <br><br>
$\delta\bar{t}=5\big(\frac{1}{5}\big)^2S^2=\frac{1}{5}S^2$
>
And so generally our <b>uncertainty in mean</b> is given by $\delta\bar{t}=\frac{1}{\sqrt{N}}S(t)$ (i.e., standard error)

<b>Additional Notes</b> <br><br>
$\;\;\;\;\;$ Systematic error dominates precision error; precision error should be as small as systematic error

```python
import numpy as np
import matplotlib
matplotlib.rcParams.update(matplotlib.rcParamsDefault)
from matplotlib import pyplot as plt

plt.rc("text", usetex=False)
plt.rc("font", family="sans-serif") #matplotlib's rcParam reset and setting usetex=False fixed a headache of an error

x, f, f1, f2, f3, f4 = np.linspace(-0.8, 0.8, 100), 1.0/np.sqrt(1-x), np.ones(len(x)), f1 + x/2, f2 + 3*x**2/8, f3 + 5*x**3/16

plt_plot_lists = [[f, "k-", r"$\frac{1}{\sqrt{1-x}}$"], [f1, "r--", r"$1$"], [f2, "b--", r"$1+\frac{1}{2}x$"], [f3, "g--", r"$1+\frac{1}{2}x+\frac{3}{8}x^2$"], [f4, "m--", r"$1+\frac{1}{2}x+\frac{3}{8}x^2+\frac{5}{16}x^3$"]]
simul_plt_plot = lambda a, b, c: plt.plot(x, a, b, label=c)
for i in plt_plot_lists:
    simul_plt_plot(i[0], i[1], i[2]) #literally did all of this to save 1 line rather than repeating plt.plot over again

plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.legend()
plt.ylim(-0.1,2.5)
plt.grid(linestyle="--")

plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from math import e

x1 = np.linspace(-1.2, 1.2, 100)
f = lambda x: e**x
plt.plot(x1, f(x1), "k-") #sample graph made, chose the simplist: e**x

tangent = lambda a, x: (e**a)*(x+1-a) #lots of iteration done to make it look simplified; basic calcI tangent line problem
a = 0.5 #choosing the point where the tangent line will be at
bounds = lambda n: (e**(1.2*(-1)**n))/(e**a)-1+a #used demos to set y=e**+/-1.2 and set them equal; a will make 1.2 +/-
x2 = np.linspace(bounds(1), bounds(2), 100)
plt.plot(x2, tangent(0.5, x2), "b-") #sample graph's tangent line at x=0.5 done

plt.plot(np.linspace(-1.2, 0.5, 100), (e**0.5)*np.ones(100), "k--") #lines for f(t_0)
plt.plot(0.5*np.ones(100), np.linspace(e**-1.2, e**0.5, 100), "k--")
plt.plot(np.linspace(-1.2, 2.5/(e**0.5)-1+a, 100), 2.5*np.ones(100), "b--") #lines for f(t_0)+delta_t
plt.plot((2.5/(e**0.5)-1+0.5)*np.ones(100), np.linspace(e**-1.2, 2.5, 100), "b--")

plt.xticks((0.5, 2.5/(e**0.5)-1+0.5), (r"$t_0$", r"$t_0+\delta t$"))
plt.yticks((2.5-e**0.5, e**0.5, 2.5), (r"$f(t_0)-\delta f$", r"$f(t_0)$", r"$f(t_0)+\delta f$"))
plt.show()
```

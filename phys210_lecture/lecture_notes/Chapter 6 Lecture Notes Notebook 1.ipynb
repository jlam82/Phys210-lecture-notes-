# <u>Chapter 6: Approximation</u>

* We have data at some (unequally spaced) points positioned at $x_j$ and at time $t_j$ for $j=0,1,\ldots,n-1$. <br><br>
* We want $v_j=\dfrac{dx_j}{dt_j}$ at other locations. <br><br>
* Need to approximate function $p(t)$.

### <u>Problems:</u>
* Approximate $f(x)$ given selected $f(x_j)$ values. <br><br>
* Typically choose $n$ basis functions $\phi_k(x)$ with a set of $n$ unknown parameters $c_k$ for $k=0,1,\ldots,n-1$. <br>
* Linear form: $p(x)=\displaystyle{ \sum_{k=0}^{n-1} }c_k\phi_k(x)$ <br>
    - $\phi_k(x)$ do not have to be linear functions, but $p(x)$ and $\phi_k(x)$ must apply to any $x$.

### <u>Interpolation:</u>
* Given $(x_j,y_j)$ for $j=0,1,\ldots,n-1$, use $\phi_k(x)$ to determine $n$ parameters of $c_k$. <br><br>
* $n$ unknowns and $n$ data points: <br><br>
* We enforce that $p(x)$ go through exactly all the input data points.
$\renewcommand{\indent}{\;\;\;\;\;}$

$$p(x_j)=y_j=\sum_{k=0}^{n-1}c_k\phi_k(x_j)$$

$$\begin{bmatrix}
    \phi_0(x_0) & \phi_1(x_0) & \phi_2(x_0) & \ldots & \phi_{n-1}(x_0) \\
    \phi_0(x_1) & \phi_1(x_1) & \phi_2(x_1) & \ldots & \phi_{n-1}(x_1) \\
    \phi_0(x_2) & \phi_1(x_2) & \phi_2(x_2) & \ldots & \phi_{n-1}(x_2) \\
    \vdots & \vdots & \vdots & \ddots &\vdots \\
    \phi_0(x_{n-1}) & \phi_1(x_{n-1}) & \phi_2(x_{n-1}) & \ldots & \phi_{n-1}(x_{n-1})
\end{bmatrix}
\begin{pmatrix}
    c_0 \\
    c_1 \\
    c_2 \\
    \vdots \\
    c_{n-1}
\end{pmatrix}
=
\begin{pmatrix}
    y_0 \\
    y_1 \\
    y_2 \\
    \vdots \\
    y_{n-1}
\end{pmatrix}$$

$\indent$ i.e., $\bf{\Phi}$$\vec{c}=\vec{y}$, where $\bf{\Phi}$ is an $n\times n$ matrix and we're solving for the $n\times 1$ column vector $\vec{c}$.

* Types or choices for $\phi(x)$: <br><br>
    - Polynomial Interpolation: assume a single polynomial can efficiently and effectively capture the behavior of the underlying function. <br><br>
    - Piecewise Polynomial Interpolation: break up data points into subintervals and employ different low-degree polynomials into each subinterval. <br><br>
    - Trigonometric Interpolation: use of trigonometric functions to interpolate periodic data. <br>
$\renewcommand{\indent}{\;\;\;\;\;}$
* Chebyshev Nodes:
$$x_j=-\cos\left(j\frac{\pi}{n-1}\right), \indent j=0,1,\ldots,n-1$$ <br>
    - Projection along the x-axis of equidistant points on the unit circle, i.e., $x_j\in[-1,1]$.

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(111)
j = np.arange(0,15,1)
f = lambda j: -np.cos(j*(np.pi/(j.shape[0]-1)))
xs = f(j)
g = lambda x: np.sqrt(1-x**2)
ys = g(xs)
ax.scatter(xs, 0*np.ones(xs.shape[0]), edgecolors="k")
ax.scatter(xs, ys, c="c", edgecolors="k")
ax.plot(np.linspace(-1.25,1.25,100), 0*np.ones(100))
ax.plot(np.linspace(-1,1,300), g(np.linspace(-1,1,300)), c="c")
ax.set_xlabel("$\mathscr{x}$")
ax.set_ylabel("$\mathscr{y}$")
```

$\;\;\;\;\;$ points cluster at the ends of the interval.

### <u>Polynomial Basis Functions:</u>
$\newcommand{\br}{\; \\}$
$$\begin{gather}
    \phi_k(x)=x^k \\
    \br
    \{1,x,x^2,\ldots,x^{n-1}\} \\
    \br
    p(x)=c_0+c_1x+c_2x^2+c_3x^3+\ldots+c_{n-1}x^{n-1}
\end{gather}$$

Evaluate at our $x_j$,
    $$p(x_j)=y_j=c_0+c_1x_j+\ldots+c_{n-1}x_j^{n-1}$$ <br>
and solve for the $c$ paramaters as $\vec{c}$.
$$\begin{pmatrix}
    1 & x_0 & x_0^2 & \ldots & x_0^{n-1} \\
    1 & x_1 & x_1^2 & \ldots & x_1^{n-1} \\
    1 & x_2 & x_2^2 & \ldots & x_2^{n-1} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    1 & x_{n-1} & x_{n-1}^2 & \ldots & x_{n-1}^{n-1}
\end{pmatrix}
\begin{pmatrix}
    c_0 \\
    c_1 \\
    c_2 \\
    \vdots \\
    c_{n-1}
\end{pmatrix} =
\begin{pmatrix}
    y_0 \\
    y_1 \\
    y_2 \\
    \vdots \\
    y_{n-1}
\end{pmatrix}$$

This (non-symmetric) $n\times n$ coefficient matrix is known as a <i>Vandermonde matrix</i>.
* The determinant of a Vandermonde matrix is non-zero for distinct nodes $x_j$. <br><br>
* Thus, a Vandermonde matrix is non-singular and $\exists$ a unique solution for $\vec{c}$.

$\underline{\mathcal{Ex:}}$ Consider the function $f(x)=\frac{1}{1+25x^2}$ (i.e., <i>Runge's function</i>). Choose $n=101$.

```python
import numpy as np
from matplotlib import pyplot as py
from math import pi

def rungefcn(x):
    return 1./(1+25*x**2)

# monomial takes dataxs, datays where function is known
# and returns interpolated values at x locations
def monomial(dataxs, datays, x):
    n = dataxs.size
    poly = np.identity(n)
    coeffs = np.ones(n)
    for i in range(n):
        poly[:,i] = dataxs**i
    coeffs = np.linalg.solve(poly,datays)
    
    p = np.ones(x.size)
    for i in range(x.size):
        xv = np.array( [x[i]**j for j in range(n)] )
        p[i] = coeffs@xv
    return p

fig = py.figure(figsize=(11,4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

n = 101
a = -1
b = 1
# get the chebyshev nodes
dataxs = -np.cos( np.linspace(0,pi,n) )
dataxs = 0.5*(a+b)+0.5*(b-a)*dataxs
datays = rungefcn(dataxs)

# we evaluate polynomial for some range of x-values interpxs
interpxs = np.linspace(a,b,100*n)
polyint = monomial(dataxs,datays,interpxs)

# plot the runge function and its interpolation (full-range)
ax1.plot(dataxs,datays,'ko', markersize=5)
ax1.plot(interpxs,polyint,'r--',linewidth=1.5)

# same plot for last 45% of values
ax2.plot(dataxs,datays,'ko')
ax2.plot(interpxs,polyint,'r--',linewidth=1.5)
#py.plot(dataxs[68:],datays[68:],'ko')
#py.plot(interpxs[76*100:],polyint[76*100:],'r-')
ax2.set_xlim(0.85,1.005)
ax2.set_ylim(0.035,0.055)
```

Some issues are present for large $n$, as seen on the ends, due to the polynomial basis functions being monomials.

### <u>Lagrange Polynomials</u>:
$\newcommand{\displayprod}[2]{ \overset{#1}{\underset{#2}{\prod}} }$
$$L_k(x)=\displaystyle{
    \frac{
        \displayprod{n-1}{j=0,j\neq k}(x-x_j)
    }{
        \displayprod{n-1}{j=0,j\neq k}(x_k-x_j)
    }
        }, \indent k=0,1,\ldots,n-1$$
* Notice: <br><br>
    - The denominator depends only on the interpolation points $x_j$. <br><br>
    - The numerator is a polynomial of $x$ with a degree of $n-1$. <br><br>
    - $L_k(x)$ has zeroes at $x_j$ when $j\neq k$.

$\underline{\mathcal{Ex:}}$ For $n=3$ (i.e., $L_k(x)$ as a quadratic), <br>

$$\begin{array}{}
    L_0(x)=\dfrac{(x-x_1)(x-x_2)}{(x_0-x_1)(x_0-x_2)} &
    L_1(x)=\dfrac{(x-x_0)(x-x_2)}{(x_1-x_0)(x_1-x_2)} &
    L_2(x)=\dfrac{(x-x_0)(x-x_1)}{(x_2-x_0)(x_2-x_1)}
\end{array}$$

When passing $x_j$'s into $L_k(x)$, observe: <br>

$$\begin{array}{}
    L_0(x_0)=1 &
    L_0(x_1)=0 &
    L_0(x_2)=0
\end{array}$$

This general result is enough to lead into <br>

$$L_k(x_j)=\delta_{kj}\text{, where }\delta_{kj}=\left\{
    \begin{array}{}
        1 \; \text{ if }k=j \\
        0 \; \text{ if }k\neq j
    \end{array}
\right.$$

We can now use $L_k(x)$ as our basis functions: <br>

$$p(x)=\sum_{k=0}^{n-1}c_k\phi_k(x)=\sum_{k=0}^{n-1}c_kL_k(x)$$

But we require $p(x_j)=y_j$ by definition: <br>

$$p(x_j)=\sum_{k=0}^{n-1}c_kL_k(x_j)=\sum_{k=0}^{n-1}c_k\delta_{kj}=c_j=y_j$$

So, $p(x)=\displaystyle{ \sum_{k=0}^{n-1} }y_kL_k(x)$

```python
import numpy as np
from matplotlib import pyplot as py

fig = py.figure(figsize=(11,4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

def nodexs(n):
    xs = -np.cos(np.linspace(0,np.pi,n))
    return xs

def LagPoly(xinter, xnodes, n, k):
    num,den = 1., 1.
    for i in range(n):
        if i == k:
            continue
        num *= (xinter - xnodes[i])
        den *= (xnodes[k] - xnodes[i])
    return num/den

npt = 100
xinter = np.linspace(-1,1,npt)

n = 3
xs = nodexs(n)
L = np.zeros( (n,xinter.size) )
for i in range(n):
    L[i] = LagPoly(xinter, xs, n, i)

ax1.plot(xinter,L[0],'k--',label=r'$L_0$')
ax1.plot(xinter,L[1],'m:', label=r'$L_1$')
ax1.plot(xinter,L[2],'c-.',label=r'$L_2$')
ax1.set_title("Lagrange Polynomials for $n=3$")
ax1.legend()

n_tilde = 6
xs = nodexs(n_tilde)
L = np.zeros( (n_tilde,xinter.size) )
for i in range(n_tilde):
    L[i] = LagPoly(xinter, xs, n_tilde, i)

ax2.plot(xinter,L[0],'k--',label=r'$L_0$')
ax2.plot(xinter,L[1],'m:', label=r'$L_1$')
ax2.plot(xinter,L[2],'c-.',label=r'$L_2$')
ax2.plot(xinter,L[3],'s', markersize=5, label=r'$L_3$')
ax2.plot(xinter,L[4],'o', markersize=5, label=r'$L_4$')
ax2.plot(xinter,L[5],'d', markersize=5, label=r'$L_5$')
ax2.set_ylim(-1,1.5)
ax2.set_title("Lagrange Polynomials for $n=6$")
ax2.legend(ncol=2)
```
